### A. **Criar o ambiente virtual (venv):**
No PowerShell, navegue até a pasta do seu projeto e crie o ambiente virtual:
```powershell
# Navegue até a pasta do projeto (substitua pelo caminho correto)
cd C:\caminho\para\seu\projeto

# Crie o ambiente virtual
python -m venv venv
```

---

### B. **Ativar o ambiente virtual no PowerShell:**
No Windows, o comando para ativar o `venv` no PowerShell é:
```powershell
.\venv\Scripts\Activate
```

- Após executar esse comando, você verá que o prompt do PowerShell mudará, mostrando o nome do ambiente virtual (`(venv)` no início da linha). Isso indica que o `venv` está ativo.

---


### **Passo 1: Verificar se o perfil foi criado corretamente**
1. Abra o PowerShell.
2. Verifique se o arquivo de perfil foi criado:
   ```powershell
   Test-Path $PROFILE
   ```
   - Se retornar `True`, o arquivo de perfil existe.
   - Se retornar `False`, o arquivo de perfil não foi criado.

3. Se o arquivo não existir, crie-o novamente:
   ```powershell
   New-Item -ItemType File -Path $PROFILE -Force
   ```

---

### **Passo 2: Editar o perfil corretamente**
1. Abra o arquivo de perfil no bloco de notas:
   ```powershell
   notepad $PROFILE
   ```

2. Adicione a função `ActivateVenv` ao arquivo. Certifique-se de que o conteúdo esteja assim:
   ```powershell
   function ActivateVenv {
       cd C:\Users\dino.grejo\Documents\Py
       .\venv\Scripts\Activate
   }
   ```

3. Salve o arquivo e feche o bloco de notas.

---

### **Passo 3: Carregar o perfil**
Após editar o perfil, você precisa carregá-lo no PowerShell para que a função `ActivateVenv` esteja disponível.

1. No PowerShell, execute:
   ```powershell
   . $PROFILE
   ```
   Isso carrega o perfil atual.

2. Agora, tente usar o comando `ActivateVenv`:
   ```powershell
   ActivateVenv
   ```

---

### **Passo 4: Verificar se o ambiente virtual foi ativado**
Se tudo estiver certo, o prompt do PowerShell deve mostrar o nome do ambiente virtual (`(venv)`) no início da linha.
