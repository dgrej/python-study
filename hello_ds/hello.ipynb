## Resumo do Fórum - Tratamento de Erros e Implementação de Modelos (Titanic Dataset)

## 1. Problemas Enfrentados e Soluções

### 1.1 Erro: `NameError: name 'x_train' is not defined`
**Causa:** Variável não definida antes do uso
**Solução:** Executar células na ordem correta e garantir a divisão dos dados

### 1.2 Erro: `FileNotFoundError: No such file or directory: 'titanic.csv'`
**Causa:** Arquivo não encontrado no diretório
**Solução:** Usar nome correto do arquivo (`titanic3.csv`)

### 1.3 Erro: `ValueError: could not convert string to float: 'male'`
**Causa:** Dados textuais sendo passados para o StandardScaler
**Solução:** Codificar variáveis categóricas (LabelEncoder) e remover colunas de texto

### 1.4 Erro: `KeyError: "['relatives'] not in index"`
**Causa:** Coluna inexistente no DataFrame
**Solução:** Verificar colunas disponíveis ou criar a partir de `sibsp` + `parch`

### 1.5 Erro: Incompatibilidade de dimensões no Keras
**Causa:** `input_dim=5` no modelo, mas dados com 4 features
**Solução:** Alterar para `input_dim=4`

### 1.6 Warning: `FutureWarning` do pandas (inplace method)
**Causa:** Método obsoleto do pandas
**Solução:** Substituir `df['col'].fillna(valor, inplace=True)` por `df['col'] = df['col'].fillna(valor)`

### 1.7 Warning: `UserWarning: Do not pass an input_shape/input_dim argument`
**Causa:** Sintaxe antiga do Keras
**Solução:** Usar `input_shape=(4,)` em vez de `input_dim=4`
